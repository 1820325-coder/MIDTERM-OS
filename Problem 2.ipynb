```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Define the System Transfer Function
num = [1]
den = [1, 2, 3, 2, 0] # Coeffs for s^4, s^3, s^2, s^1, s^0
sys = ctrl.TransferFunction(num, den)
```

```python
# 2. Compute Step Response
# Note: Pole at s=0 indicates an integrator.
t, y = ctrl.step_response(sys)
```

```python
# 3. Visualization
plt.figure(figsize=(14, 6))
```

```python
# Subplot 1: Step Response Plot
plt.subplot(1, 2, 1)
plt.plot(t, y, 'b', label='Displacement X2(t)')
plt.title('System Step Response')
plt.grid(True)
```

```python
# Subplot 2: Pole-Zero Map
plt.subplot(1, 2, 2)
ctrl.pole_zero_plot(sys)
plt.title('System Pole-Zero Map')
plt.grid(True)

plt.tight_layout()
plt.show()
```
