```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define coefficients
num = [2, 1, 3]
den = [8, 12, 26, 18, 0]
```

```python
# System definition
system = ct.TransferFunction(num, den)
```

```python
# Transient analysis
time, response = ct.step_response(system)
```

```python
# Visualization
plt.figure(figsize=(10,6))
plt.plot(time, response, color='#00838F', linewidth=2)
plt.title('Step Response Analysis', fontsize=14, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Displacement X2(s)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Stability check
plt.figure()
ct.pole_zero_plot(system)
plt.title('Pole-Zero Map')
plt.grid(True)
plt.show()
```
