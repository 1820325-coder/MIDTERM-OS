```python
import control
import matplotlib.pyplot as plt
import numpy as np

# Define the Transfer Function
# Numerator: s + 1 (Coefficients: [1, 1])
# Denominator: s^4 + 4s^3 + 7s^2 + 6s + 3 (Coefficients: [1, 4, 7, 6, 3])
num = [1.0, 1.0]
den = [1.0, 4.0, 7.0, 6.0, 3.0]
sys = control.TransferFunction(num, den)

print("Symbolic Transfer Function H(s):")
print(sys)
```

```python
# 1. Step Response Plot
# This plot is used to evaluate transient behavior like overshoot and settling time.
time, response = control.step_response(sys)
plt.figure()
plt.plot(time, response)
plt.title('Step Response: Transient Behavior Evaluation')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement X2')
plt.grid(True)
```

```python
# 2. Pole-Zero Map
# This plot is used to verify system stability based on the location of the poles.
plt.figure()
control.pzmap(sys, plot=True)
plt.title('Pole-Zero Map: Stability Verification')
plt.grid(True)

# Display plots
plt.show()
```
